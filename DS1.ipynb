# Task
Create a simple expert system capable of making basic decisions based on predefined rules, including defining its scope, designing its knowledge base, implementing an inference engine, and testing its decision-making capabilities.

## Define Expert System Scope

### Subtask:
Establish the specific domain for the expert system's simple decision-making. For example, it could be a system to recommend a drink based on temperature and mood, or diagnose a simple issue based on symptoms.


### Chosen Expert System Domain: Drink Recommendation

**Description:**
This expert system will recommend a drink based on two primary factors: the user's current mood and the outdoor temperature. The goal is to provide a simple, personalized drink suggestion to enhance the user's experience.

**Specific Decisions/Recommendations:**
The system will recommend a specific type of drink (e.g., 'Hot Coffee', 'Iced Tea', 'Lemonade', 'Hot Chocolate') that is appropriate for the given mood and temperature.

**Key Inputs:**
*   `temperature`: An integer representing the outdoor temperature (e.g., in Celsius or Fahrenheit).
*   `mood`: A string describing the user's current mood (e.g., 'Happy', 'Sad', 'Tired', 'Energetic').

**Possible Outputs:**
*   `recommendation`: A string suggesting a drink (e.g., 'Hot Coffee', 'Iced Tea').

## Design Knowledge Base

### Subtask:
Outline the structure for storing rules and facts, representing the knowledge base using Python data structures.


### Temperature and Mood Categories

To effectively categorize user input for drink recommendations, we will define specific ranges for temperature and a set of recognizable moods.

#### Temperature Categories:
*   **Cold**: Temperatures below 10°C.
*   **Moderate**: Temperatures between 10°C and 25°C (inclusive).
*   **Hot**: Temperatures above 25°C.

#### Mood Categories:
*   **Happy**: Reflects a cheerful and positive state.
*   **Sad**: Indicates a feeling of sorrow or unhappiness.
*   **Tired**: Suggests a need for energy or relaxation due to fatigue.
*   **Energetic**: Represents high spirits and readiness for activity.
*   **Relaxed**: Denotes a calm and peaceful state.

**Reasoning**:
Now that the categories are defined, I will create a Python dictionary to represent the knowledge base, mapping combinations of mood and temperature categories to drink recommendations, and populate it with example rules.



```python
knowledge_base = {
    ('Happy', 'Hot'): 'Iced Tea',
    ('Happy', 'Moderate'): 'Lemonade',
    ('Happy', 'Cold'): 'Hot Chocolate',
    ('Sad', 'Hot'): 'Mint Tea',
    ('Sad', 'Moderate'): 'Warm Milk with Honey',
    ('Sad', 'Cold'): 'Coffee',
    ('Tired', 'Hot'): 'Iced Coffee',
    ('Tired', 'Moderate'): 'Espresso',
    ('Tired', 'Cold'): 'Energy Drink',
    ('Energetic', 'Hot'): 'Sparkling Water with Lime',
    ('Energetic', 'Moderate'): 'Green Smoothie',
    ('Energetic', 'Cold'): 'Fresh Juice',
    ('Relaxed', 'Hot'): 'Herbal Infusion',
    ('Relaxed', 'Moderate'): 'Chamomile Tea',
    ('Relaxed', 'Cold'): 'Still Water'
}

print("Knowledge Base Structure (Mood, Temperature) -> Drink Recommendation:")
for (mood, temp), drink in knowledge_base.items():
    print(f"  If mood is '{mood}' and temperature is '{temp}', recommend: {drink}")
```

## Implement Inference Engine

### Subtask:
Develop the core logic that processes user inputs against the knowledge base to derive a decision or recommendation.


**Reasoning**:
I need to define two functions: `categorize_temperature` to classify the temperature into predefined categories and `get_drink_recommendation` to process user input (temperature and mood) and retrieve a drink recommendation from the `knowledge_base` based on the defined rules.



```python
def categorize_temperature(temperature):
    if temperature < 10:
        return 'Cold'
    elif 10 <= temperature <= 25:
        return 'Moderate'
    else:
        return 'Hot'

def get_drink_recommendation(user_temperature, user_mood, knowledge_base):
    # Ensure mood is capitalized for consistency with knowledge base
    user_mood_capitalized = user_mood.capitalize()

    # Categorize the user_temperature
    temperature_category = categorize_temperature(user_temperature)

    # Check if mood is valid (optional, but good for robust systems)
    valid_moods = {'Happy', 'Sad', 'Tired', 'Energetic', 'Relaxed'}
    if user_mood_capitalized not in valid_moods:
        return f"Sorry, '{user_mood}' is not a recognized mood. Please choose from {', '.join(valid_moods)}."

    # Construct the query tuple
    query = (user_mood_capitalized, temperature_category)

    # Look up the recommendation in the knowledge base
    recommendation = knowledge_base.get(query, "No specific recommendation for this input.")

    return recommendation

print("Functions `categorize_temperature` and `get_drink_recommendation` defined.")
```

## Test the Expert System

### Subtask:
Provide example inputs to test the expert system's decision-making capabilities and verify that it produces the expected outputs based on its rules.


**Reasoning**:
To test the expert system, I will create a series of test cases with different temperatures and moods, including one with an unrecognized mood, and then call the `get_drink_recommendation` function for each case, printing the results to verify its functionality.



```python
print("\n--- Testing Expert System ---")

# Test Cases
test_cases = [
    (28, 'Happy'),      # Hot, Happy -> Iced Tea
    (15, 'Tired'),      # Moderate, Tired -> Espresso
    (5, 'Sad'),         # Cold, Sad -> Coffee
    (30, 'Energetic'),  # Hot, Energetic -> Sparkling Water with Lime
    (18, 'Relaxed'),    # Moderate, Relaxed -> Chamomile Tea
    (22, 'sleepy')      # Unrecognized mood
]

for temp, mood in test_cases:
    recommendation = get_drink_recommendation(temp, mood, knowledge_base)
    print(f"For temperature {temp}°C and mood '{mood}', the recommendation is: {recommendation}")
```
